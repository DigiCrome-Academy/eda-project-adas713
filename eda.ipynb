## Rename Columns

```python
df = df.rename(columns={'race/ethnicity':'ethnic_group', 'parental level of education':'parent_education', 'test preparation course':'test_prep', 'math score': 'math_score', 'reading score': 'reading_score', 'writing score': 'writing_score'})
```

## Boxplot

```python
fig, axes = plt.subplots(1, 3, figsize = (15,5))

sns.boxplot(x='gender', y='math_score', data=df, ax=axes[0])
axes[0].set_title("Math Scores by Gender")

sns.boxplot(x='gender', y='reading_score', data=df, ax=axes[1])
axes[1].set_title("Reading Scores by Gender")

sns.boxplot(x='gender', y='writing_score', data=df, ax=axes[2])
axes[2].set_title("Writing Scores by Gender")
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15,5))

sns.boxplot(x='ethnic_group', y='math_score', data=df, ax=axes[0])
axes[0].set_title("Math Scores by Ethnic Group")

sns.boxplot(x='ethnic_group', y='reading_score', data=df, ax=axes[1])
axes[1].set_title("Reading Scores by Ethnic Group")

sns.boxplot(x='ethnic_group', y='writing_score', data=df, ax=axes[2])
axes[2].set_title("Writing Scores by Ethnic Group")
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15,5))

sns.boxplot(x='parent_education', y='math_score', data=df, ax=axes[0])
axes[0].set_title("Math Scores by Parent Education Level")
axes[0].tick_params(axis='x', rotation=90)

sns.boxplot(x='parent_education', y='reading_score', data=df, ax=axes[1])
axes[1].set_title("Reading Scores by Parent Education Level")
axes[1].tick_params(axis='x', rotation=90)

sns.boxplot(x='parent_education', y='writing_score', data=df, ax=axes[2])
axes[2].set_title("Writing Scores by Parent Education Level")
axes[2].tick_params(axis='x', rotation=90)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15,5))

sns.boxplot(x='lunch', y='math_score', data=df, ax=axes[0])
axes[0].set_title("Math Scores by Lunch Status")

sns.boxplot(x='lunch', y='reading_score', data=df, ax=axes[1])
axes[1].set_title("Reading Scores by Lunch Status")

sns.boxplot(x='lunch', y='writing_score', data=df, ax=axes[2])
axes[2].set_title("Writing Scores by Lunch Status")
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15,5))

sns.boxplot(x='test_prep', y='math_score', data=df, ax=axes[0])
axes[0].set_title("Math Scores by Test Preparation Status")

sns.boxplot(x='test_prep', y='reading_score', data=df, ax=axes[1])
axes[1].set_title("Reading Scores by Test Preparation Status")

sns.boxplot(x='test_prep', y='writing_score', data=df, ax=axes[2])
axes[2].set_title("Writing Scores by Test Preparation Status")
```

# Multivariate Analysis

## Heatmap

```python
corr_matrix = df[['math_score', 'reading_score', 'writing_score']].corr()

sns.heatmap(corr_matrix)
plt.title("Correlation Matrix of Scores")
```

## Grouped Bar Charts

```python
grouped_gender = df.groupby("gender")[['math_score', 'reading_score', 'writing_score']].mean()

grouped_gender.plot(kind="bar")
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
```

```python
grouped_ethnic = df.groupby("ethnic_group")[['math_score', 'reading_score', 'writing_score']].mean()

grouped_ethnic.plot(kind="bar")
plt.title("Average Scores by Ethnic Group")
plt.xlabel("Ethnic Group")
plt.ylabel("Average Score")
```

```python
grouped_parented = df.groupby("parent_education")[['math_score', 'reading_score', 'writing_score']].mean()

grouped_parented.plot(kind="bar")
plt.title("Average Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Average Score")
```

```python
grouped_lunch = df.groupby("lunch")[['math_score', 'reading_score', 'writing_score']].mean()

grouped_lunch.plot(kind="bar")
plt.title("Average Scores by Lunch Status")
plt.xlabel("Lunch Status")
plt.ylabel("Average Score")
```

```python
grouped_testprep = df.groupby("test_prep")[['math_score', 'reading_score', 'writing_score']].mean()

grouped_testprep.plot(kind="bar")
plt.title("Average Scores by Test Preparation Status")
plt.xlabel("Test Preparation Status")
plt.ylabel("Average Score")
```
